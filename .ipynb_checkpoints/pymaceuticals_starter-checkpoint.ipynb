```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset

combined_data = pd.merge(mouse_metadata, study_results, how = "left", on = "Mouse ID")

# Display the data table for preview
combined_data.head()

```

```python
# Checking the number of mice.
mice_number = len(combined_data["Mouse ID"].unique())
mice_number
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dup_data = (combined_data.loc[combined_data.duplicated(['Mouse ID', "Timepoint"])])
dup_data
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
clean_data = combined_data[combined_data["Mouse ID"] != 'g989']
clean_data
```

```python
# Checking the number of mice in the clean DataFrame.

mice_count = len(clean_data["Mouse ID"].unique())
mice_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

summ_mean = clean_data.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
summ_median = clean_data.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
summ_var = clean_data.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
summ_stdev = clean_data.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
summ_sem = clean_data.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']


summary_df = pd.DataFrame({"Mean": summ_mean,
                           "Median": summ_median,
                           "Variance": summ_var,
                           "Std. Dev": summ_stdev,
                           "SEM": summ_sem})

                                 
summary_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.

summary_df_line = clean_data.groupby(["Drug Regimen"])[['Tumor Volume (mm3)']].agg(["mean", "median", "var", "std", "sem"])
                                               

summary_df_line 
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

capomulin_only = clean_data.loc[clean_data["Drug Regimen"] == 'Capomulin']

#avg_tumor = pd.DataFrame(capomulin_only.groupby('Mouse ID')['Tumor Volume (mm3)'].mean().sort_values()).reset_index()
#avg_tumor
capomul_avg = capomulin_only.groupby(['Mouse ID']).mean()
capomul_avg

#capomulin_only
plt.scatter(capomul_avg['Weight (g)'], capomul_avg['Tumor Volume (mm3)'])

plt.xlabel('Weight (g)')
plt.ylabel('Avg. Tumor Volume (mm3)')

plt.show()

```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

corr = st.pearsonr(capomul_avg['Weight (g)'], capomul_avg['Tumor Volume (mm3)'])
corr
plt.scatter(capomul_avg['Weight (g)'], capomul_avg['Tumor Volume (mm3)'])

plt.xlabel('Weight (g)')
plt.ylabel('Avg. Tumor Volume (mm3)')

linear= st.linregress(capomul_avg['Weight (g)'], capomul_avg['Tumor Volume (mm3)'])


linear                                                          

#plt.show()
#corr
slope=0.9544396890241049
stderr=0.12754359033201323
intercept=21.552160532685008
intercept_stderr=2.5909876867683455
rvalue=0.8419363424694721

pe_fit = slope * (capomul_avg['Weight (g)'] + intercept)
plt.plot(capomul_avg['Weight (g)'], pe_fit, "--", color='r')


```
